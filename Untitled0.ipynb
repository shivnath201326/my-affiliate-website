```
#!/usr/bin/env python3
"""
40th-order video PDE demo (semi-implicit spectral IMEX)
Save as: video_pde_40th.py
Run:     python video_pde_40th.py

Dependencies:
  - Python 3.8+
  - PyTorch (cpu or gpu) - pip install torch
  - matplotlib - pip install matplotlib

Notes:
  - This is a toy demo for teaching / prototyping. For larger grids/long runs,
    prefer a GPU and increase system memory/timeouts accordingly.
  - If runtime is long, reduce nx/ny or steps.
"""
import math, time
import torch
import torch.fft as fft
import matplotlib.pyplot as plt

# ---------------- config (tweak these for speed/quality) ----------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
dtype = torch.float32

# Grid: choose small values to test quickly; increase for richer behavior.
nx = 128          # recommended: 64..256 depending on compute
ny = 128
Lx = 2*math.pi
Ly = 2*math.pi
dx = Lx / nx
dy = Ly / ny

# Time stepping
dt = 5e-4         # time-step (smaller if stiff or unstable)
steps = 800       # total timesteps
save_every = 200  # capture frames for visualization

# PDE params
kappa = 1e-7      # coefficient for Delta^{20} term (tiny because |k|^40 grows large)
sigma = 0.0       # stochastic forcing (0 for deterministic)
alpha_frac = 1.2  # exponent used in neural operator low-pass (mild)

# ---------------- create spectral multipliers ----------------
kx = 2*math.pi * torch.fft.fftfreq(nx, d=dx, device=device, dtype=dtype)
ky = 2*math.pi * torch.fft.rfftfreq(ny, d=dy, device=device, dtype=dtype)
Kx = kx.unsqueeze(1).repeat(1, ky.shape[0])       # shape (nx, ny//2+1)
Ky = ky.unsqueeze(0).repeat(kx.shape[0], 1)
K2 = Kx**2 + Ky**2                                # |k|^2
K2_pow20 = K2**20                                 # |k|^{40} for Delta^{20}

# ---------------- initial conditions ----------------
xx = torch.linspace(0, Lx, nx, device=device, dtype=dtype)
yy = torch.linspace(0, Ly, ny, device=device, dtype=dtype)
X, Y = torch.meshgrid(xx, yy, indexing='ij')

# U: video frame field (initial)
U0 = (0.6*torch.sin(3*X) * torch.cos(2*Y) +
      0.45*torch.exp(-((X - 2.0)**2 + (Y - 1.5)**2)/0.18) +
      0.25*torch.sin(5*X + 2*Y))
U0 = U0 - U0.mean()
U0 = U0 / (U0.abs().max() + 1e-12)

# V: latent image potential (drives advection)
V = 0.7*torch.sin(2*X - 0.5*Y) + 0.2*torch.cos(3*Y)
V = V - V.mean()
V = V / (V.abs().max() + 1e-12)

U = U0.to(device)
V = V.to(device)

# ---------------- neural operator surrogate (spectral low-pass + coupling) ----------------
def theta_model(U: torch.Tensor, V: torch.Tensor) -> torch.Tensor:
    """
    Lightweight surrogate for a learned operator.
    Performs a spectral low-pass on U and couples pointwise with V.
    """
    U_hat = fft.rfft2(U)
    # low-pass multiplier (stable, simple)
    p = 0.6 + 0.4 * alpha_frac
    lp = 1.0 / (1.0 + (K2**p))
    U_smooth = fft.irfft2(U_hat * lp, s=(nx, ny))
    coupling = 0.02 * (U_smooth * torch.tanh(V))
    return coupling

# ---------------- spectral differentiation helpers ----------------
iKx = 1j * Kx
iKy = 1j * Ky

def spectral_gradient(field: torch.Tensor):
    """Return (d/dx, d/dy) using spectral differentiation (periodic)."""
    fh = fft.rfft2(field)
    fx = fft.irfft2(fh * iKx, s=(nx, ny))
    fy = fft.irfft2(fh * iKy, s=(nx, ny))
    return fx.real, fy.real

# ---------------- semi-implicit IMEX step ----------------
def step_imex(U, V, dt, kappa, sigma, K2_pow20):
    # advective term: -div(U * grad V)
    Vx, Vy = spectral_gradient(V)
    # periodic conservative derivatives via roll (central difference)
    def dxd(f):
        return (torch.roll(f, shifts=-1, dims=0) - torch.roll(f, shifts=1, dims=0)) / (2*dx)
    def dyd(f):
        return (torch.roll(f, shifts=-1, dims=1) - torch.roll(f, shifts=1, dims=1)) / (2*dy)
    term_x = dxd(U * Vx)
    term_y = dyd(U * Vy)
    adv = -(term_x + term_y)

    Ntheta = theta_model(U, V)
    noise = sigma * torch.randn_like(U) * math.sqrt(dt) if sigma > 0 else 0.0
    NL = adv + Ntheta + noise    # nonlinear part (explicit)

    # spectral semi-implicit update (treat linear high-order term implicitly)
    U_hat = fft.rfft2(U)
    NL_hat = fft.rfft2(NL)
    denom = 1.0 + dt * kappa * K2_pow20
    U_hat_new = (U_hat + dt * NL_hat) / denom
    U_new = fft.irfft2(U_hat_new, s=(nx, ny))
    return U_new

# ---------------- run simulation ----------------
frames = []
U_curr = U.clone()
frames.append(U_curr.cpu().numpy())

print(f"Running on device: {device}, grid: {nx}x{ny}, steps: {steps}")
t0 = time.time()
for n in range(1, steps+1):
    U_curr = step_imex(U_curr, V, dt, kappa, sigma, K2_pow20)
    if n % save_every == 0 or n == steps:
        frames.append(U_curr.cpu().numpy())
        print(f"captured step {n}/{steps}  max={U_curr.max():.4e}  min={U_curr.min():.4e}")
t1 = time.time()
print(f"Done in {t1-t0:.2f}s, captured {len(frames)} frames.")

# ---------------- visualize (initial / one mid / final) ----------------
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.title("Initial U")
plt.imshow(frames[0], origin='lower', extent=(0,Lx,0,Ly))
plt.axis('off')

mid_idx = len(frames)//2
plt.subplot(1,3,2)
plt.title(f"Intermediate (idx {mid_idx})")
plt.imshow(frames[mid_idx], origin='lower', extent=(0,Lx,0,Ly))
plt.axis('off')

plt.subplot(1,3,3)
plt.title("Final U")
plt.imshow(frames[-1], origin='lower', extent=(0,Lx,0,Ly))
plt.axis('off')

plt.suptitle("40th-order video PDE â€” semi-implicit spectral demo")
plt.show()
```

```
# PyTorch-style pseudocode (requires torch >= 1.8 with complex FFT)
import torch
import torch.fft as fft

def wavevectors(nx, ny, dx, dy, device):
    kx = 2*torch.pi*torch.fft.fftfreq(nx, d=dx).to(device)
    ky = 2*torch.pi*torch.fft.fftfreq(ny, d=dy).to(device)
    Kx, Ky = torch.meshgrid(kx, ky, indexing='ij')
    K2 = Kx**2 + Ky**2
    return K2  # shape (nx,ny)

def delta20_hat(U_hat, K2):
    # (-|k|^2)^{20} U_hat = |k|^{40} * U_hat
    return (K2**20) * U_hat

# Forward step: semi-implicit IMEX
def step_IMEX(U, V, theta_model, K2, dt, kappa, device):
    # U, V are real tensors shape (nx,ny) or (C,nx,ny)
    # compute nonlinear term NL = -div(U grad V) + N_theta(U,V) + noise
    # 1) compute advective term (spatial derivatives in real domain)
    Ux = torch.gradient(U, spacing=(dx,dy), dim=( -2, -1))[0]   # sketch
    Uy = torch.gradient(U, spacing=(dx,dy), dim=( -2, -1))[1]
    Vx = torch.gradient(V, spacing=(dx,dy), dim=( -2, -1))[0]
    Vy = torch.gradient(V, spacing=(dx,dy), dim=( -2, -1))[1]
    adv = -(Ux*Vx + Uy*Vy) - U*(torch.div(Vx,dx) + torch.div(Vy,dy)) # approximate div(U grad V)
    Ntheta = theta_model(U, V)  # user-supplied neural operator
    noise = sigma * torch.randn_like(U) * torch.sqrt(dt)

    NL = adv + Ntheta + noise

    # move to Fourier
    U_hat = fft.rfft2(U)
    NL_hat = fft.rfft2(NL)

    denom = 1.0 + dt * kappa * (K2**20)  # (1 + dt * kappa * |k|^40)
    U_hat_new = (U_hat + dt * NL_hat) / denom

    U_new = fft.irfft2(U_hat_new, s=U.shape[-2:])
    return U_new
```

```
#!/usr/bin/env python3
"""
40th-order video PDE demo (semi-implicit spectral IMEX)
Save as: video_pde_40th.py
Run:     python video_pde_40th.py

Dependencies:
  - Python 3.8+
  - PyTorch (cpu or gpu) - pip install torch
  - matplotlib - pip install matplotlib

Notes:
  - This is a toy demo for teaching / prototyping. For larger grids/long runs,
    prefer a GPU and increase system memory/timeouts accordingly.
  - If runtime is long, reduce nx/ny or steps.
"""
import math, time
import torch
import torch.fft as fft
import matplotlib.pyplot as plt

# ---------------- config (tweak these for speed/quality) ----------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
dtype = torch.float32

# Grid: choose small values to test quickly; increase for richer behavior.
nx = 128          # recommended: 64..256 depending on compute
ny = 128
Lx = 2*math.pi
Ly = 2*math.pi
dx = Lx / nx
dy = Ly / ny

# Time stepping
dt = 5e-4         # time-step (smaller if stiff or unstable)
steps = 800       # total timesteps
save_every = 200  # capture frames for visualization

# PDE params
kappa = 1e-7      # coefficient for Delta^{20} term (tiny because |k|^40 grows large)
sigma = 0.0       # stochastic forcing (0 for deterministic)
alpha_frac = 1.2  # exponent used in neural operator low-pass (mild)

# ---------------- create spectral multipliers ----------------
kx = 2*math.pi * torch.fft.fftfreq(nx, d=dx, device=device, dtype=dtype)
ky = 2*math.pi * torch.fft.rfftfreq(ny, d=dy, device=device, dtype=dtype)
Kx = kx.unsqueeze(1).repeat(1, ky.shape[0])       # shape (nx, ny//2+1)
Ky = ky.unsqueeze(0).repeat(kx.shape[0], 1)
K2 = Kx**2 + Ky**2                                # |k|^2
K2_pow20 = K2**20                                 # |k|^{40} for Delta^{20}

# ---------------- initial conditions ----------------
xx = torch.linspace(0, Lx, nx, device=device, dtype=dtype)
yy = torch.linspace(0, Ly, ny, device=device, dtype=dtype)
X, Y = torch.meshgrid(xx, yy, indexing='ij')

# U: video frame field (initial)
U0 = (0.6*torch.sin(3*X) * torch.cos(2*Y) +
      0.45*torch.exp(-((X - 2.0)**2 + (Y - 1.5)**2)/0.18) +
      0.25*torch.sin(5*X + 2*Y))
U0 = U0 - U0.mean()
U0 = U0 / (U0.abs().max() + 1e-12)

# V: latent image potential (drives advection)
V = 0.7*torch.sin(2*X - 0.5*Y) + 0.2*torch.cos(3*Y)
V = V - V.mean()
V = V / (V.abs().max() + 1e-12)

U = U0.to(device)
V = V.to(device)

# ---------------- neural operator surrogate (spectral low-pass + coupling) ----------------
def theta_model(U: torch.Tensor, V: torch.Tensor) -> torch.Tensor:
    """
    Lightweight surrogate for a learned operator.
    Performs a spectral low-pass on U and couples pointwise with V.
    """
    U_hat = fft.rfft2(U)
    # low-pass multiplier (stable, simple)
    p = 0.6 + 0.4 * alpha_frac
    lp = 1.0 / (1.0 + (K2**p))
    U_smooth = fft.irfft2(U_hat * lp, s=(nx, ny))
    coupling = 0.02 * (U_smooth * torch.tanh(V))
    return coupling

# ---------------- spectral differentiation helpers ----------------
iKx = 1j * Kx
iKy = 1j * Ky

def spectral_gradient(field: torch.Tensor):
    """Return (d/dx, d/dy) using spectral differentiation (periodic)."""
    fh = fft.rfft2(field)
    fx = fft.irfft2(fh * iKx, s=(nx, ny))
    fy = fft.irfft2(fh * iKy, s=(nx, ny))
    return fx.real, fy.real

# ---------------- semi-implicit IMEX step ----------------
def step_imex(U, V, dt, kappa, sigma, K2_pow20):
    # advective term: -div(U * grad V)
    Vx, Vy = spectral_gradient(V)
    # periodic conservative derivatives via roll (central difference)
    def dxd(f):
        return (torch.roll(f, shifts=-1, dims=0) - torch.roll(f, shifts=1, dims=0)) / (2*dx)
    def dyd(f):
        return (torch.roll(f, shifts=-1, dims=1) - torch.roll(f, shifts=1, dims=1)) / (2*dy)
    term_x = dxd(U * Vx)
    term_y = dyd(U * Vy)
    adv = -(term_x + term_y)

    Ntheta = theta_model(U, V)
    noise = sigma * torch.randn_like(U) * math.sqrt(dt) if sigma > 0 else 0.0
    NL = adv + Ntheta + noise    # nonlinear part (explicit)

    # spectral semi-implicit update (treat linear high-order term implicitly)
    U_hat = fft.rfft2(U)
    NL_hat = fft.rfft2(NL)
    denom = 1.0 + dt * kappa * K2_pow20
    U_hat_new = (U_hat + dt * NL_hat) / denom
    U_new = fft.irfft2(U_hat_new, s=(nx, ny))
    return U_new

# ---------------- run simulation ----------------
frames = []
U_curr = U.clone()
frames.append(U_curr.cpu().numpy())

print(f"Running on device: {device}, grid: {nx}x{ny}, steps: {steps}")
t0 = time.time()
for n in range(1, steps+1):
    U_curr = step_imex(U_curr, V, dt, kappa, sigma, K2_pow20)
    if n % save_every == 0 or n == steps:
        frames.append(U_curr.cpu().numpy())
        print(f"captured step {n}/{steps}  max={U_curr.max():.4e}  min={U_curr.min():.4e}")
t1 = time.time()
print(f"Done in {t1-t0:.2f}s, captured {len(frames)} frames.")

# ---------------- visualize (initial / one mid / final) ----------------
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.title("Initial U")
plt.imshow(frames[0], origin='lower', extent=(0,Lx,0,Ly))
plt.axis('off')

mid_idx = len(frames)//2
plt.subplot(1,3,2)
plt.title(f"Intermediate (idx {mid_idx})")
plt.imshow(frames[mid_idx], origin='lower', extent=(0,Lx,0,Ly))
plt.axis('off')

plt.subplot(1,3,3)
plt.title("Final U")
plt.imshow(frames[-1], origin='lower', extent=(0,Lx,0,Ly))
plt.axis('off')

plt.suptitle("40th-order video PDE â€” semi-implicit spectral demo")
plt.show()
```
